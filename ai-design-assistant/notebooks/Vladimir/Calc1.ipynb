### Текстовые функции (СЖПРОБЕЛЫ, ПРОПНАЧ, ЛЕВСИМВ, ПРАВСИМВ...)

2. ПРОПНАЧ - Для работы с текстом, первая буква становится заглавной, остальные строчные
3. ЛЕВСИМВ - Позволяет извлечь определенное количество символов с левой стороны текста
4. ПРАВСИМВ - Позволяет извлечь определенное количество символов с правой стороны текста

### Работа со строками (СЦЕП, ОБЪЕДИНИТЬ, &)

1. СЦЕП - обьеденяет несколько ячеек в одну =СЦЕП(A1; " "; B1) где А1 и В1 ячейки для обьединения, в ковычках пробел, но может быть любой знак препинания
2. ОБЪЕДИНИТЬ - ттоже обьединяет ячейки в одну, больше подходит для создания заголовков, информации, которую потом можно выделить, отформатировать..
3. & - соединяет значения ячеек или строки с определенной ячейкой  =A1&" "&B1
