## Регулярные выражения

Регулярные выражения (RegEx) — инструмент для поиска и обработки текста по шаблону.

### Поиск конкретных символов

- кот — найдет "кот" в слове "котенок".
- 123 — найдет "123"

### Спецсимволы (Метасимволы)

Примеры:
- \d — любая цифра (от 0 до 9).
- \D — любой символ, кроме цифры.
- \w — любая буква (латиница), цифра или нижнее подчеркивание.
- \s — любой пробельный символ (пробел, табуляция).
- . (точка) — вообще любой один символ (кроме переноса строки).

### Квантификаторы (Счётчики)

- \* — 0 или более раз.
- \+ — 1 или более раз.
- ? — 0 или 1 раз.
- {n} — ровно n раз.
- {n,m} — от n до m раз.

### Границы и группы

- ^ — начало строки.
- $ — конец строки.
- \[abc] — любой из указанных символов (в данном случае 'a', 'b' или 'c').
- | — логическое «ИЛИ».
- () — группа

### Практика

1. *Текст:* "В 2024 на 50 %, а в 2025 уже на 170 %." *Задача:* Найти все года.- *\d\d\d\d    \d{4}*
2. *Текст:* Валидация формы регистрации на сайтах *Задача:* Логин должен начинаться с буквы, содержать только латинские буквы и цифры, быть длиной от 3 до 10 символов. *[a-zA-Z][a-zA-Z\d]{2,9}  ^([a-zA-Z][a-zA-Z\d]{2,9})$*
3. *Задача:* Найти номера, начинающиеся с +7 или 8 и содержащие 11 цифр  *\d{11}     ^[+]|\d{11}*
4. *Текст:* "«ivan@mail.ru», (petr@gmail.com), \[anna@yandex.ru]" *Задача:* Оставить только чистый адрес (без скобок и кавычек).  *Найти: [^a-zA-Z0-9.@]  Заменить:(поставить пробел)  На пробел заменятся все символы, за исключением перечисленных в квадратных скобках*
5. *Текст:* "Товар стоит 1500р., доставка 300 руб, скидка 50 р" *Задача:* Вытащить все суммы (только числа), после которых идет упоминание рублей в любом виде.    *\d{1,4}   \d{1,4}([ руб]{4}|[руб]{3}|[ р]{2}|[р]{1})*
6. *Текст:* "2023-12-31" *Задача:* Превратить в формат 31.12.2023  *Формат ячейки- формат даты- выбрать нужное из списка форматирования*
