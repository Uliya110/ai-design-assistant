## 1/10 Генерация нормального распределения

```python
import numpy as np
# TODO: сгенерируйте 1000 значений нормального распределения
data = np.random.normal(0, 1, 1000)
len(data)
```

## 2/10 Среднее и стандартное отклонение

```python
# TODO: вычислите mean и std
mean = np.mean(data)
std = np.std(data)
mean, std
```

## 3/10 Гистограмма нормального распределения

```python
import matplotlib.pyplot as plt
# TODO: постройте гистограмму data
plt.hist(data, bins=30)
plt.title('Normal Distribution')
plt.show()
```

## 4/10 Добавление выброса

```python
# TODO: добавьте выброс 10
data_out = np.append(data, 10)
len(data_out)
```

## 5/10 Гистограмма с выбросом

```python
# TODO: постройте гистограмму data_out
plt.hist(data_out, bins=30)
plt.title('With Outlier')
plt.show()
```

## 6/10 Медиана

```python
# TODO: вычислите медиану
median = np.median(data_out)
median
```

## 7/10 Проверка правила 68%

```python
# TODO: реализуйте проверку 68%
lower = mean - std
upper = mean + std
within = np.sum((data >= lower) & (data <= upper))
within / len(data)
```

## 8/10 Равномерное распределение

```python
# TODO: сгенерируйте uniform и постройте гистограмму
uniform = np.random.uniform(-3, 3, 1000)
plt.hist(uniform, bins=30)
plt.title('Uniform Distribution')
plt.show()
```

## 9/10 Сравнение средних

```python
# TODO: сравните средние
np.mean(data), np.mean(uniform)
```
