### Что важно понять
Текст в Python хранится в типе данных **`str`**.  
Со строками можно:

- получать длину текста
- извлекать отдельные символы
- брать фрагменты строки
- очищать лишние пробелы
- менять регистр
- заменять символы
- объединять несколько строк в одну

### Базовые термины
- **Строка** — текстовое значение
- **Индекс** — номер символа в строке
- **Срез** — часть строки
- **Метод строки** — встроенная операция, которую можно применить к тексту

## Практическая ячейка 2. Индексация и срезы

### Что делаем
Получаем отдельные символы и части строки.

### Функции, методы и синтаксис
- `строка[0]` — первый символ
- `строка[-1]` — последний символ
- `строка[0:2]` — срез
- `строка[-4:]` — последние 4 символа

### Задание
1. Найдите первый символ.
2. Найдите последний символ.
3. Получите префикс `AI`.
4. Получите последние 4 символа года.

```python
product_code = "DI-CENTRE-2026"

# TODO:
first_symbol = product_code[0]
last_symbol = product_code[-1]
prefix = product_code[0:2]
year_part = product_code[-4:]

print("Первый символ:", first_symbol)
print("Последний символ:", last_symbol)
print("Префикс:", prefix)
print("Год:", year_part)
```

## Практическая ячейка 3. Очистка текста и методы строк

### Что делаем
Очищаем строку от лишних пробелов, меняем регистр и заменяем символы.

### Функции, методы и синтаксис
- `.strip()`
- `.lower()`
- `.upper()`
- `.replace("старое", "новое")`

### Задание
1. Уберите пробелы с краёв.
2. Переведите строку в нижний регистр.
3. Переведите строку в верхний регистр.
4. Замените `-` на пробел.

```python
raw_category = "  Аналитика-ДАННЫХ  "

# TODO:
clean_category = raw_category.strip()
lower_category = clean_category.lower()
upper_category = clean_category.upper()
replaced_category = lower_category.replace("-", " ")

print("Исходная строка:", repr(raw_category))
print("После strip():", repr(clean_category))
print("После lower():", lower_category)
print("После upper():", upper_category)
print("После replace():", replaced_category)
```

## Практическая ячейка 4. Склейка строк и форматирование текста

### Что делаем
Объединяем несколько текстовых частей в одно значение.

### Функции, методы и синтаксис
- `+` — склейка
- `f"..."` — форматированная строка

### Задание
1. Соберите полное имя.
2. Сформируйте e-mail в нижнем регистре.
3. Сделайте красивую подпись с отделом.

```python
first_name = "Светлана"
last_name = "Окс"
department = "DI Lab"

# TODO:
full_name = first_name + " " + last_name
email = first_name.lower() + "." + last_name.lower() + "@example.com"
label = f"{full_name} | отдел: {department}"

print("Полное имя:", full_name)
print("E-mail:", email)
print("Подпись:", label)
```

## Практическая ячейка 5. Мини-задача: очистка карточки товара

### Что делаем
Берём «грязные» текстовые данные и превращаем их в аккуратные значения для анализа.

### Функции, методы и синтаксис
- `.strip()`
- `.lower()`
- `.upper()`
- `.replace()`
- `f"..."`

### Задание
1. Очистите название товара.
2. Сделайте бренд в верхнем регистре.
3. Сделайте категорию в нижнем регистре.
4. Соберите итоговую карточку товара.

```python
raw_product_name = "  Компьютерное_кресло  "
raw_brand = "  FrandeR "
raw_category = "  ОФИСНАЯ МЕБЕЛЬ "

# TODO:
product_name = raw_product_name.strip().replace("_", " ")
brand = raw_brand.strip().upper()
category = raw_category.strip().lower()
card_label = f"{brand} | {product_name} | категория: {category}"

print("Очищенное название товара:", product_name)
print("Бренд:", brand)
print("Категория:", category)
print("Итоговая карточка:", card_label)
```
