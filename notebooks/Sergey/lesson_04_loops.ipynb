# Занятие 4. Циклы и повторяющиеся расчёты

## Краткая теория по ноутбуку

На этом занятии мы учимся автоматизировать повторяющиеся действия.

В аналитике постоянно встречаются задачи, где одно и то же действие нужно выполнить:
- для каждой строки
- для каждого заказа
- для каждого клиента
- для каждого значения в списке

В Excel для этого часто копируют формулу вниз.  
В Python для этого используют **циклы**.

### Что важно понять
Цикл позволяет:
- перебрать числа по порядку
- пройти по списку значений
- накопить итог
- посчитать количество элементов по условию

### Базовые термины
- **Цикл** — повторение одного и того же блока кода
- **Итерация** — один шаг цикла
- **Счётчик** — переменная, которая увеличивается
- **Накопление** — поэтапное суммирование значений

### Что изучаем сегодня
1. Цикл `for` и `range()`
2. Перебор списка значений
3. Накопление суммы
4. Подсчёт количества элементов по условию
5. Мини-задачу на анализ продаж

### Зачем это нужно разработчику систем с ИИ
До сложных моделей почти всегда есть этап обычной обработки данных:
- пройти по записям
- проверить значения
- собрать агрегаты
- отметить проблемные элементы
- получить итоговые показатели

Циклы — это базовая автоматизация таких шагов.


## Практическая ячейка 1. Цикл `for` и `range()`

### Что делаем
Повторяем действие несколько раз и смотрим на номер шага.

### Функции, методы и синтаксис
- `for ... in ...:` — цикл
- `range(1, 6)` — последовательность чисел от 1 до 5
- `print(...)` — вывод результата

### На что обратить внимание
Каждый проход цикла — это одна итерация.  
`range(1, 6)` не включает последнее значение 6.


```python
for day in range(1, 6):
    print(f"День {day}: обработка данных завершена")
```

## Практическая ячейка 2. Цикл по списку значений

### Что делаем
Перебираем список продаж и выводим каждое значение.

### Функции, методы и синтаксис
- `список = [...]` — создание списка
- `for item in список:` — перебор элементов списка
- `print(...)` — вывод текущего элемента

### На что обратить внимание
Цикл берёт элементы списка по одному и позволяет выполнять с ними одинаковые действия.


```python
sales = [1200, 1500, 980, 2100, 1750]

for sale in sales:
    print("Продажа за день:", sale)
```

## Практическая ячейка 3. Накопление суммы

### Что делаем
Считаем общую сумму продаж через цикл.

### Функции, методы и синтаксис
- `total = 0` — начальное значение накопителя
- `total = total + value` — прибавление текущего элемента
- сокращённая запись: `total += value`

### На что обратить внимание
Накопитель нужно создать до начала цикла.


```python
sales = [1200, 1500, 980, 2100, 1750]
total_sales = 0

for sale in sales:
    total_sales += sale

print("Общая сумма продаж:", total_sales)
```

## Практическая ячейка 4. Подсчёт количества элементов по условию

### Что делаем
Считаем, сколько продаж больше заданного порога.

### Функции, методы и синтаксис
- `if условие:` — проверка условия
- счётчик обычно начинается с нуля
- если условие выполнено, счётчик увеличивается на 1

### На что обратить внимание
Это базовый приём для анализа данных: сколько заказов, клиентов или записей удовлетворяют правилу.


```python
sales = [1200, 1500, 980, 2100, 1750]
threshold = 1500
count_above_threshold = 0

for sale in sales:
    if sale > threshold:
        count_above_threshold += 1

print("Порог:", threshold)
print("Количество продаж выше порога:", count_above_threshold)
```

## Практическая ячейка 5. Мини-задача: анализ списка продаж

### Что делаем
Собираем несколько расчётов в одной задаче:
- общую сумму
- количество продаж
- среднюю продажу
- число дней выше порога

### Функции, методы и синтаксис
- цикл `for`
- накопление суммы
- счётчик элементов
- проверка условия через `if`

### На что обратить внимание
Это уже мини-анализ данных, очень близкий к реальной практике.


```python
sales = [1200, 1500, 980, 2100, 1750, 1900, 1600]
threshold = 1500

total_sales = 0
days_count = 0
days_above_threshold = 0

for sale in sales:
    total_sales += sale
    days_count += 1
    if sale > threshold:
        days_above_threshold += 1

average_sale = total_sales / days_count

print("Список продаж:", sales)
print("Общая сумма:", total_sales)
print("Количество дней:", days_count)
print("Средняя продажа:", round(average_sale, 2))
print("Дней выше порога:", days_above_threshold)
```

## Ячейка 6. Тест и самопроверка

### Что делаем
Проверяем:
- цикл `for`
- `range()`
- перебор списка
- накопление суммы
- подсчёт по условию

### Функции, методы и синтаксис
- `assert условие` — быстрая проверка результата


```python
# 1. range()
numbers = []
for n in range(1, 4):
    numbers.append(n)
assert numbers == [1, 2, 3]

# 2. Перебор списка и сумма
values = [10, 20, 30]
total = 0
for value in values:
    total += value
assert total == 60

# 3. Подсчёт элементов по условию
sales = [900, 1600, 1700, 800]
count_high = 0
for sale in sales:
    if sale > 1000:
        count_high += 1
assert count_high == 2

# 4. Мини-проверка среднего
sales = [1000, 2000, 3000]
total_sales = 0
days_count = 0
for sale in sales:
    total_sales += sale
    days_count += 1
average_sale = total_sales / days_count
assert average_sale == 2000

print("Тест пройден успешно. Все проверки выполнены.")
```
