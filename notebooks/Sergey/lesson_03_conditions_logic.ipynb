# Занятие 3. Условия, логика и принятие решений в Python и Excel

## Краткая теория по ноутбуку

На этом занятии мы учимся принимать решения на основе данных.  
Именно это делает аналитику, автоматизацию и многие системы с ИИ полезными для бизнеса.

Когда программа видит число, статус, флаг или текстовое значение, она может:

- сравнить значение с порогом
- проверить, выполняется ли условие
- выбрать один из нескольких сценариев
- присвоить статус, скидку или категорию

### Что важно понять
В Python для логики используются:

- операторы сравнения: `>`, `<`, `>=`, `<=`, `==`, `!=`
- логический тип `bool`
- конструкции `if`, `elif`, `else`
- логические операторы `and`, `or`, `not`

В Excel аналогично используются:

- сравнения в формулах
- функция `IF`
- связки `AND`, `OR`, `NOT`

### Базовые термины
- **Условие** — проверка, которая даёт результат `True` или `False`
- **Логическое выражение** — выражение со сравнением или логическими связками
- **Ветка решения** — действие, которое выполняется при истинности или ложности условия
- **Порог** — значение, относительно которого принимается решение

### Что изучаем сегодня
1. Простые сравнения и логические результаты
2. Конструкцию `if / else`
3. Несколько веток через `if / elif / else`
4. Сложные условия через `and`, `or`, `not`
5. Мини-задачу, похожую на реальную обработку заявки

### Зачем это нужно разработчику систем с ИИ
Перед тем как использовать модели ИИ, данные и процессы часто проходят через обычную бизнес-логику:
- кому дать скидку
- какой заказ считать срочным
- какую заявку отправить менеджеру
- когда показать предупреждение
- когда пропустить запись дальше по конвейеру


## Практическая ячейка 1. Сравнения и логические значения

### Что делаем
Сравниваем числовые значения и получаем результаты `True` или `False`.

### Функции, методы и синтаксис
- `>` — больше
- `<` — меньше
- `>=` — больше или равно
- `<=` — меньше или равно
- `==` — равно
- `!=` — не равно
- `print(...)` — вывод результата

### На что обратить внимание
Результат сравнения — это логическое значение `True` или `False`.
Именно такие результаты потом используются внутри `if`.


```python
order_amount = 12500
free_delivery_threshold = 10000
discount_percent = 10

print("Сумма заказа:", order_amount)
print("Порог бесплатной доставки:", free_delivery_threshold)
print("Сумма заказа больше порога?", order_amount > free_delivery_threshold)
print("Сумма заказа равна порогу?", order_amount == free_delivery_threshold)
print("Скидка не равна нулю?", discount_percent != 0)
print("Скидка не больше 15%?", discount_percent <= 15)
```

## Практическая ячейка 2. Простое условие `if / else`

### Что делаем
Проверяем, положена ли клиенту бесплатная доставка.

### Функции, методы и синтаксис
- `if условие:` — выполнить блок, если условие истинно
- `else:` — выполнить альтернативный блок
- отступ после `if` обязателен

### На что обратить внимание
В Python блоки кода определяются отступами.  
Если забыть отступ, программа выдаст ошибку.


```python
order_amount = 8700
free_delivery_threshold = 10000

if order_amount >= free_delivery_threshold:
    delivery_status = "Бесплатная доставка"
else:
    delivery_status = "Платная доставка"

print("Сумма заказа:", order_amount)
print("Статус доставки:", delivery_status)
```

## Практическая ячейка 3. Несколько веток: `if / elif / else`

### Что делаем
Определяем уровень скидки по сумме заказа.

### Функции, методы и синтаксис
- `if` — первая проверка
- `elif` — следующая проверка, если предыдущая не сработала
- `else` — сценарий по умолчанию

### На что обратить внимание
Условия проверяются сверху вниз.  
Как только одно условие сработало, остальные уже не проверяются.


```python
order_amount = 18500

if order_amount >= 20000:
    discount_level = "Скидка 15%"
elif order_amount >= 10000:
    discount_level = "Скидка 10%"
elif order_amount >= 5000:
    discount_level = "Скидка 5%"
else:
    discount_level = "Скидки нет"

print("Сумма заказа:", order_amount)
print("Уровень скидки:", discount_level)
```

## Практическая ячейка 4. Сложные условия: `and`, `or`, `not`

### Что делаем
Определяем, нужно ли передать заказ на ручную проверку.

### Функции, методы и синтаксис
- `and` — оба условия должны быть истинными
- `or` — хотя бы одно условие должно быть истинным
- `not` — логическое отрицание

### На что обратить внимание
Сложные условия помогают описывать реальные бизнес-правила:
например, высокий чек и новый клиент, или долг и срочная доставка.


```python
order_amount = 26000
is_new_client = True
has_debt = False
express_delivery = True

manual_review = (order_amount > 20000 and is_new_client) or (has_debt or express_delivery)

print("Сумма заказа:", order_amount)
print("Новый клиент:", is_new_client)
print("Есть задолженность:", has_debt)
print("Срочная доставка:", express_delivery)
print("Нужна ручная проверка?", manual_review)
print("Клиент без долга?", not has_debt)
```

## Практическая ячейка 5. Мини-задача: обработка заявки клиента

### Что делаем
На основе нескольких признаков определяем итоговый статус заявки.

### Функции, методы и синтаксис
- сравнения для чисел и логических флагов
- `if / elif / else` для выбора статуса
- `and` и `not` для комбинирования правил

### На что обратить внимание
Это уже очень похоже на реальную задачу автоматизации:
мы не просто сравниваем значения, а принимаем решение по записи.


```python
request_amount = 14000
documents_ready = True
manager_approved = True
client_blacklisted = False

if client_blacklisted:
    request_status = "Отклонить заявку"
elif documents_ready and manager_approved and request_amount <= 15000:
    request_status = "Одобрить автоматически"
elif documents_ready and not manager_approved:
    request_status = "Отправить менеджеру"
else:
    request_status = "Запросить документы"

print("Сумма заявки:", request_amount)
print("Документы готовы:", documents_ready)
print("Менеджер одобрил:", manager_approved)
print("Клиент в чёрном списке:", client_blacklisted)
print("Итоговый статус:", request_status)
```

## Ячейка 6. Тест и самопроверка

### Что делаем
Проверяем:
- сравнения
- `if / else`
- `if / elif / else`
- логические операторы `and`, `or`, `not`
- принятие решения по нескольким условиям

### Функции, методы и синтаксис
- `assert условие` — быстрая проверка правильности результата


```python
# 1. Простое сравнение
amount = 12000
assert amount > 10000

# 2. if / else
score = 78
if score >= 60:
    exam_status = "Сдано"
else:
    exam_status = "Не сдано"
assert exam_status == "Сдано"

# 3. if / elif / else
purchase = 7200
if purchase >= 10000:
    bonus = "Высокий"
elif purchase >= 5000:
    bonus = "Средний"
else:
    bonus = "Низкий"
assert bonus == "Средний"

# 4. and / or / not
is_vip = True
has_overdue = False
priority = is_vip and not has_overdue
assert priority is True

# 5. Комплексное правило
documents_ready = True
manager_approved = False
blacklisted = False

if blacklisted:
    status = "Отклонить"
elif documents_ready and manager_approved:
    status = "Одобрить"
elif documents_ready and not manager_approved:
    status = "Проверить вручную"
else:
    status = "Запросить документы"

assert status == "Проверить вручную"

print("Тест пройден успешно. Все проверки выполнены.")
```
